## 2. Download Script
The following Bash script will download the [Smart Meters In London](https://www.kaggle.com/jeanmidev/smart-meters-in-london) dataset from Kaggle.

```python
%%bash

#Create a subfolder to store the data
mkdir data_1
cd data_1

#Download the data
kaggle datasets download -d jeanmidev/smart-meters-in-london

#Unzip the downloaded data
unzip smart-meters-in-london.zip

#Unzip any files which were zip files
unzip hhblock_dataset.zip
unzip daily_dataset.zip
unzip halfhourly_dataset.zip

#Remove the zip files (but keep the original download)
rm hhblock_dataset.zip
rm halfhourly_dataset.zip
rm daily_dataset.zip
```
